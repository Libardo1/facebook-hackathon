```python
import numpy as np
import utils
from keras.models import load_model
from keras import backend as K

from matplotlib import pyplot as plt
from data import Dataset, DataLoader
from model import IMG_WIDTH, IMG_HEIGHT
%matplotlib inline  

model = load_model('../models/2017-10-07-13-09-51.hdf5')
model.summary()
ds = Dataset()
X, y = ds.get_test_data()
gen = DataLoader(X, y, IMG_WIDTH, IMG_HEIGHT)
```

```python
X, y = gen[1]
seed_img = np.expand_dims(X[np.random.randint(0, len(X))], 0)
compute_fn = K.function([model.input, K.learning_phase()], 
                        [model.layers[-9].output, model.layers[-2].output, model.output])
mask, refined_mask, pred = compute_fn([seed_img, 0])

print("True: {}, Pred {}".format(y[0], pred[0]))
plt.imshow(utils.decode(seed_img[0]))
plt.show()

plt.imshow(mask[0, ..., 0] / np.max(mask), cmap='gray')
plt.colorbar()
plt.show()

plt.imshow(refined_mask[0, ..., 0] / np.max(refined_mask), cmap='gray')
plt.colorbar()
plt.show()
```
